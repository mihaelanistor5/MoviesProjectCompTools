```python
import os
reviewed_movie = os.listdir('/content/drive/MyDrive/DTU/Computational Tools for Data Science/project/dataset/2_reviews_per_movie_raw')
# print("The number of reviewed movies:{}".format(len(reviewed_movie)))
reviewed_movie
```

```python
# For removing .csv for each component in the list
movie_name = []
for movie in reviewed_movie:
    if movie.count(".") == 1: 
        name = movie.split('.')[0]
        movie_name.append(name)
    else:
        for i in range(len(movie)-1,0,-1):
            if movie[i]=='.':
                movie_name.append(movie[:i])
                break
                
movie_name
```
