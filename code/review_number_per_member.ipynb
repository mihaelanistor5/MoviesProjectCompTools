```python
cnt = 0
df_fin = pd.DataFrame()
for files in folder:

  df = pd.read_csv(files)
  
  list_movie = []
  for j in range(len(df)):
    list_movie.append(movie_reviewed[cnt])

  df.insert(0,'movie_name',list_movie)
  df = df.drop(['username'],axis=1)
  rating_list = []
  
  helpful = []

  total = []

  date = []

  title = []

  review = []

  for i in df['rating']:
    rating_list.append(i)

  for h in df['helpful']:
    helpful.append(h)
  
  for j in df['total']:
    total.append(j)

  for k in df['date']:
    date.append(k)

  for l in df['title']:
    title.append(l)

  for m in df['review']:
    review.append(m)
  
  row = []
  row.append(movie_reviewed[cnt])
  row.append(rating_list)
  row.append(helpful)
  row.append(total)
  row.append(date)
  row.append(title)
  row.append(review)

  df_tem = pd.DataFrame(columns=['movie_name','rating','helpful','total','date','title','review'])
  df_tem = df_tem.append(pd.Series(row, index=df.columns), ignore_index=True)

  df_fin = pd.concat([df_fin,df_tem])
  cnt+=1
 
df_fin = df_fin.reset_index()
df_fin
```
