```python
from sklearn.feature_extraction.text import CountVectorizer

# identify it by the space(' ')
df_cp_2['genres_literal'] = df_cp_2['genres'].apply(lambda x: (' ').join(x))
count_vect = CountVectorizer(min_df=0, ngram_range=(1,2))
genre_mat = count_vect.fit_transform(df_cp_2['genres_literal'])
```

```python
from sklearn.metrics.pairwise import cosine_similarity

genre_sim = cosine_similarity(genre_mat, genre_mat)
genre_sim
```

```python
genre_sim_sorted = genre_sim.argsort()[:,: :-1]
genre_sim_sorted
```

```python
def find_sim_movie(df,sorted,movie_name,top_n=5):
  title_movie = df[df['original_title']==movie_name]
  idx = title_movie.index.values
  similar_idx = sorted[idx, :(top_n)]

  # print(similar_idx)
  similar_idx = similar_idx.reshape(-1)

  return df.iloc[similar_idx]

similar_movies = find_sim_movie(df_cp_2, genre_sim_sorted,'Avatar',5)
similar_movies[['original_title','vote_average']]
```

## Using weighted rating - ex) Recommend movies to people who watched the movie "Avatar"

```python
# Weighted Rating formula in IMDB
# Weighted Rating = v/(v+m))* R + (m/(v+m))*C(
# v: vote_count, m: the minimum number for giving rate
# R: the average rate of each movie -> vote_average, C: the average rate of all movies
percentage = 0.6
C = df_cp_2['vote_average'].mean()
m = df_cp_2['vote_count'].quantile(percentage)

def weighted_vote_avg(record):
  v = record['vote_count']
  R = record['vote_average']
  
  formula = ((v/(v+m))* R + (m/(v+m))*C)
  return formula

df_cp_2['weighted_vote'] = df_cp_2.apply(weighted_vote_avg,axis=1)
df_cp_2[['original_title','vote_average','weighted_vote','vote_count']].sort_values('weighted_vote',ascending=False)[:5]
```

```python
def find_sim_movie_weighted(df, sorted,title_name,top_n=5):
  title_movie = df[df['original_title']==title_name]
  title_idx = title_movie.index.values

  similar_idx = sorted[title_idx,:(top_n*2)]
  similar_idx = similar_idx.reshape(-1)

  similar_idx = similar_idx[similar_idx!=title_idx]

  return df.iloc[similar_idx].sort_values('weighted_vote',ascending=False)[:top_n]

similar_weighted = find_sim_movie_weighted(df_cp_2, genre_sim_sorted,'Avatar',5)
similar_weighted[['original_title','vote_average','weighted_vote']]
```
