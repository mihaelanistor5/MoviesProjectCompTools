# Item-based CF

## Rating

```python
final_cc = final.copy()
tmp = final_cc['username'].value_counts()[-4878:]
tmp_ = tmp.index.tolist()

list_tmp = []

for i in range(len(final_cc)):
    tm = final_cc.iloc[i]
    if tm['username'] in tmp_:
        list_tmp.append(i)

final_cc = final_cc.drop(list_tmp)
final_cc
```

```python
ratings = final_cc.copy()

ratings = ratings[['original_title','username','ratings']]

def RMSE(y_true,y_pred):
  for i in range(len(y_true)):
    tmp = y_true[i]
    f_tmp = float(tmp)
    y_true[i] = f_tmp
  
  for i in range(len(y_pred)):
    if y_pred[i] == np.nan:
      y_pred[i] = 0.0
    else:
      tmp = y_pred[i]
      f_tmp = float(tmp)
      y_pred[i] = f_tmp
  
  return np.sqrt(np.mean((np.array(y_true)-np.array(y_pred))**2))

# def score(model,neighbor_size=0): # neighbor size!
#   id_pairs = zip(x_test['username'],x_test['original_title'])
#   y_pred = np.array([model(user,movie,neighbor_size) for (user,movie) in id_pairs])
#   y_true = np.array(x_test['rating'])

#   return RMSE(y_true,y_pred) 

x = ratings.copy()
y = ratings['username']

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3,stratify=y)

rating_matrix = x_train.pivot(index='username',columns = 'original_title',values = 'ratings')
rating_matrix
```

```python
rating_matrix_t = np.transpose(rating_matrix)

matrix_dummy = rating_matrix_t.copy().fillna(0)
item_similarity = cosine_similarity(matrix_dummy, matrix_dummy)
item_similarity = pd.DataFrame(item_similarity,
                               index = rating_matrix_t.index,
                               columns = rating_matrix_t.index)
```

```python
def score_(model):
  id_pairs = zip(x_test['username'],x_test['original_title'])
  y_pred = np.array([model(user,movie) for (user,movie) in id_pairs])
  y_true = np.array(x_test['ratings'])

  return RMSE(y_true,y_pred)

def CF_IBCF(user_id,movie_id):
  if movie_id in item_similarity.columns:
    sim_scores = item_similarity[movie_id]
    user_rating = rating_matrix_t[user_id]
    # the indexes of not rated movies 
    none_rating_idx = user_rating[user_rating.isnull()].index
    # exclude nan
    user_rating = user_rating.dropna()
    for i in range(len(user_rating)):
      tmp = user_rating[i]
      int_tmp = int(tmp)
      user_rating[i] = int_tmp
    # exclude not rated ones
    sim_scores = sim_scores.drop(none_rating_idx)
    if sim_scores.sum()==0:
      mean_rating = 5.45
    # mean_rating = np.dot(sim_scores,user_rating) / sim_scores.sum()
    else:
      mean_rating = np.dot(sim_scores,user_rating) / sim_scores.sum()

  else:
    mean_rating = 5.45
  return mean_rating

score_(CF_IBCF)
```

```python
def recom_movie(user_id,n_items):
  user_movie = rating_matrix.loc[user_id].copy()

  for movie in rating_matrix.columns: 
    if pd.notnull(user_movie.loc[movie]): 
      user_movie.loc[movie] = 0
    else:
      user_movie.loc[movie] = CF_IBCF(user_id,movie)
  movie_sort = user_movie.sort_values(ascending=False)[:n_items]
  
  recom_movies = movie_sort.index.tolist()
  
  return recom_movies

ratings_recom = recom_movie(user_id = 'action-man101', n_items=5) 
ratings_recom
```
